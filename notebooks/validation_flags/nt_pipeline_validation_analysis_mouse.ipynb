```python
from riot_na.data.model import Scheme
from pathlib import Path
import pandas as pd
from IPython.display import Markdown
from notebooks.utils import base64_decode_series, calculate_validation_flags_summary
```

# Validation flags results - Nucleotides, mouse

```python
RESULTS_DIR = Path().absolute().parent.parent / "results" / "mouse_database_nt"
```

```python
for scheme in Scheme:
    path = RESULTS_DIR / f"ngs_sample_clean_{scheme.value}_mouse.csv"
    df = pd.read_csv(
        path,
        index_col=0,
        engine="pyarrow",
        usecols=["sequence_header", "v_call", "productive", "additional_validation_flags"],
    )
    df = base64_decode_series(df, "additional_validation_flags")
    display(Markdown(f"## {scheme.upper()}"))
    display(calculate_validation_flags_summary(df))
    display(Markdown("---"))
```
